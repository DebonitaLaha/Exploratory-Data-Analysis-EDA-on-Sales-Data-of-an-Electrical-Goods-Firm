### Data Cleaning

#### Identification of NaN rows

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()
```

#### Dropping NaN 

```python
all_data = all_data.dropna(how = 'all')
all_data.head()
```

#### Identifying anything other than the date 

```python
all_data.head()
```

```python
temp_data = all_data [all_data['Order Date'].str[0:2] == 'Or'] 
temp_data.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or'] #removes the duplicating rows
temp_data.head()
```

### Augmenting Data 

#### Adding columns for the month from Order Date  

```python
all_data['Month'] = all_data ['Order Date'].str[0:2]
all_data['Month'] = all_data ['Month'].astype('int32')
all_data.head()
```

### Analysis  

#### Q1: Best month for Sales

##### Adding a Sales column 

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data ['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data ['Price Each'])


all_data['Sales'] = all_data ['Quantity Ordered'] * all_data ['Price Each']
all_data.head()
```

##### Sorting by Months

```python
all_data.groupby('Month').sum()
```

##### Visualization

```python
months = range (1,13)
add_result = all_data.groupby('Month').sum()
plt.bar(months , add_result['Sales'] , color='blue')
plt.xticks(months)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

#### Q2: City with highest Sales?

##### Spliting the city name

```python
def get_city(adress):
    return adress.split(',')[1]  #spliting the city names according the ',' and extracting the first index
```

##### Spliting the State along with City

```python
def get_state(adress):
    return adress.split(',')[2].split(' ')[1]
```

```python
all_data['City'] = all_data['Purchase Address'].apply (lambda x: get_city(x)+ ' ('+ get_state(x)+')')
all_data.head()
```

##### Sum of sales in each States

```python
 all_data.groupby('City').sum()

```

##### Visualisation

```python
#city_name = all_data['City'].unique()  # doesnot get the data in order same as the group by values

city_name = [city for city, df in all_data.groupby('City')] #makes the data in order of the group by

add_result = all_data.groupby('City').sum()

plt.bar(city_name, add_result['Sales'] , color='red')
plt.xticks(city_name , rotation='vertical',size=8)
plt.xlabel("City Name" )
plt.ylabel("Sales")
plt.show()
```

#### Q3: Best time to display advertisement to maximize the likelihood of Customer's buying of product.

##### Conversion of the date-time into date Time Object

```python
all_data ['Order Date'] = pd.to_datetime(all_data ['Order Date'])
```

##### Adding Hour and Minute columns

```python
all_data ['Hour'] = all_data ['Order Date'].dt.hour
all_data ['Minute'] = all_data ['Order Date'].dt.minute
all_data.head()
```

```python
unique_hour = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(all_data.groupby('Hour').count())
plt.xticks(unique_hour , rotation='vertical',size=8)
plt.xlabel("Hours" )
plt.ylabel("Sales in each Hour")
plt.grid()
plt.show()
```

#### Q4: Products which are sold together quite often?

##### Creating a separte dataset consisting only duplicates

```python
df = all_data[all_data['Order Date'].duplicated(keep=False)] #keep=false to keep all occurance of the duplicate

df['Grouped Products'] = df.groupby('Order ID')['Product']. transform(lambda x: ','.join(x)) #Grouping of the products with similar order ID

df= df[['Order ID','Grouped Products']].drop_duplicates()


df.head()
```

##### Products bought together the most

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped Products']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2))) #3 for top 3 items bought together
    
    
count.most_common(10)    
```
